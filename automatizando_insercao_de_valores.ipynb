````python
from openpyxl import load_workbook, Workbook
````

- **openpyxl**: Permite que você leia e grave arquivos do Excel nos formatos modernos baseados em XML (.xlsx, .xlsm).
- **defusedxml**: Para aumentar a segurança ao trabalhar com arquivos de fontes não confiáveis.

> Para referenciar células, use ```sheet["A1"]```

## Carregar dados

````python
wb = load_workbook('ATTOMI 19-01.xlsx', read_only=False, data_only=False)
sheet = wb.active
````

- **load_workbook**: Carrega o arquivo excel
  - **read_only**: parâmetro booleano que prende somente o código a somente a leitura do arquivo excel, isso é bom performaticamente, mas limita os recursos de edição.
  - **data_only**: parâmetro booleano que se for True retorna valores calculados em vez de fórmulas.

## Iterar dados

````python
# Iterando as linhas
for row in sheet.iter_rows(min_row=1, max_row=5, values_only=True):
    print(row)

# Iterando as colunas
for column in sheet.iter_cols(min_row=1, max_row=5, values_only=True):
    print(column)

range = sheet["A3:O45"]
for row in range:
  for cell in row:
    print(cell.value)
````
