```python
import yt_dlp
```

```python
def extraindo_links(playlist):
  # Extrae apenas metadados
  ydl_opts = {
      'extract_flat': True, # Não baixa os vídeos
      'quiet': True, # Terminal fica menos poluído
      'ignorerrors': True, # Pula vídeos removidos/privados
  }

  links = [] # lista aonde os links ficarão armazenados

  print("Iniciando operação, isso pode levar alguns minutos...")

  with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    try:
      # Extrai informações da playlist
      informacoes_dict = ydl.extract_info(playlist, download=False)
      if 'entries' in informacoes_dict:
        # Itera sobre os vídeos encontrados
        for video in informacoes_dict['entries']:
          link = f"https://www.youtube.com/watch?v={video.get('id')}"
          links.append(link)
          print(f"Encontrado: {video.get('title')}")

      return links
    except Exception as e:
      print(f"Ocorreu um erro: {e}")
      return []

# -----Configurações------
url_da_playlist = input("Cole a URL da playlist aqui:")
links_extraidos = extraindo_links(url_da_playlist)

# Salvar em arquivo
if links_extraidos:
    with open("links_videos.txt", "w", encoding="utf-8") as f:
        for link in links_extraidos:
            f.write(link + "\n")
    print(f"\nSucesso! {len(links_extraidos)} links salvos em 'links_videos.txt'.")
else:
    print("Nenhum link encontrado.")
```
