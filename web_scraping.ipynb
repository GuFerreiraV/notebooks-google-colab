```python
import time
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
```

```python
def get_all_react_links(base_url):
    response = requests.get(base_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    links = set()

    for a_tag in soup.find_all('a', href=True):
        href = a_tag['href']
        # Converte links relativos em absolutos
        full_url = urljoin(base_url, href)
        # Filtra para garantir que permanece no domínio react.dev
        if urlparse(full_url).netloc == 'react.dev':
            links.add(full_url)

    return links

# Execução
react_links = get_all_react_links("https://react.dev/learn")
for link in sorted(react_links):
    print(link)
    time.sleep(2)
```

## Web Scraping - ASP.NET

```python
def get_all_aspnet_links(base_url):
  response = requests.get(base_url)
  soup = BeautifulSoup(response.text, 'html.parser')
  links = set()

  for a_tag in soup.find_all('a', href=True):
    href = a_tag['href']
    full_url = urljoin(base_url, href)
    if urlparse(full_url).netloc == 'learn.microsoft.com':
      links.add(full_url)

  return links

aspnet_links = get_all_aspnet_links("https://learn.microsoft.com/pt-br/aspnet/mvc/overview/")
for link in sorted(aspnet_links):
  print(link)
  time.sleep(2)
```
