# Jogo da Velha (Tic-Tac-Toe)

```python
def criar_tabuleiro():
    return ['1', '2', '3', '4', '5', '6', '7', '8', '9']

tabuleiro = criar_tabuleiro()
print("Tabuleiro inicial:")
print(tabuleiro)
```

```python
def exibir_tabuleiro(tabuleiro):
    print(f'\n{tabuleiro[0]} | {tabuleiro[1]} | {tabuleiro[2]}')
    print('---------')
    print(f'{tabuleiro[3]} | {tabuleiro[4]} | {tabuleiro[5]}')
    print('---------')
    print(f'{tabuleiro[6]} | {tabuleiro[7]} | {tabuleiro[8]}\n')

exibir_tabuleiro(tabuleiro)
```

```python
def verificar_vitoria(tabuleiro, jogador):
    # Linhas
    if (tabuleiro[0] == jogador and tabuleiro[1] == jogador and tabuleiro[2] == jogador) or \
       (tabuleiro[3] == jogador and tabuleiro[4] == jogador and tabuleiro[5] == jogador) or \
       (tabuleiro[6] == jogador and tabuleiro[7] == jogador and tabuleiro[8] == jogador):
        return True
    # Colunas
    if (tabuleiro[0] == jogador and tabuleiro[3] == jogador and tabuleiro[6] == jogador) or \
       (tabuleiro[1] == jogador and tabuleiro[4] == jogador and tabuleiro[7] == jogador) or \
       (tabuleiro[2] == jogador and tabuleiro[5] == jogador and tabuleiro[8] == jogador):
        return True
    # Diagonais
    if (tabuleiro[0] == jogador and tabuleiro[4] == jogador and tabuleiro[8] == jogador) or \
       (tabuleiro[2] == jogador and tabuleiro[4] == jogador and tabuleiro[6] == jogador):
        return True
    return False
```

```python
def verificar_empate(tabuleiro):
    for casa in tabuleiro:
        if casa in ['1', '2', '3', '4', '5', '6', '7', '8', '9']:
            return False # Ainda há casas vazias
    return True # Todas as casas preenchidas e ninguém venceu
```

```python
def jogar_jogo_da_velha():
    while True:
        tabuleiro = criar_tabuleiro()
        jogador_atual = 'X'
        jogo_ativo = True

        print("\n--- Início do Jogo da Velha ---")

        while jogo_ativo:
            exibir_tabuleiro(tabuleiro)
            print(f"É a vez do jogador {jogador_atual}.")

            try:
                posicao = int(input("Escolha uma posição (1-9): "))
                if 1 <= posicao <= 9 and tabuleiro[posicao - 1] not in ['X', 'O']:
                    tabuleiro[posicao - 1] = jogador_atual
                else:
                    print("Posição inválida ou já ocupada. Tente novamente.")
                    continue
            except ValueError:
                print("Entrada inválida. Por favor, digite um número entre 1 e 9.")
                continue

            if verificar_vitoria(tabuleiro, jogador_atual):
                exibir_tabuleiro(tabuleiro)
                print(f"Parabéns! O jogador {jogador_atual} venceu!")
                jogo_ativo = False
            elif verificar_empate(tabuleiro):
                exibir_tabuleiro(tabuleiro)
                print("O jogo terminou em empate!")
                jogo_ativo = False
            else:
                jogador_atual = 'O' if jogador_atual == 'X' else 'X'

        jogar_novamente = input("Quer jogar novamente? (s/n): ").lower()
        if jogar_novamente != 's':
            break

    print("Obrigado por jogar!")
```

Para iniciar o jogo, execute a célula abaixo:

```python
jogar_jogo_da_velha()
```
