### 3. Cálculo do Valor de Venda com Lucro
Crie um programa que permita digitar o nome do produto e o valor de compra, e imprima o nome do produto e o valor da venda, aplicando a tabela de lucros fornecida.

```python
def calcular_valor_venda():
    nome_produto = input("Digite o nome do produto: ")
    valor_compra = float(input("Digite o valor de compra do produto: R$ "))

    if valor_compra < 10:
        lucro_percentual = 0.70
    elif 10 <= valor_compra < 30:
        lucro_percentual = 0.50
    elif 30 <= valor_compra < 50:
        lucro_percentual = 0.40
    else:
        lucro_percentual = 0.30

    valor_venda = valor_compra * (1 + lucro_percentual)

    print(f"\nNome do Produto: {nome_produto}")
    print(f"Valor de Venda: R$ {valor_venda:.2f}")

calcular_valor_venda()
```

### 4. Calculadora Simples
Elabore um programa em Python que implemente uma calculadora com as funções de somar, subtrair, multiplicar e dividir. O programa deverá solicitar ao usuário os dois valores, e perguntar qual a operação pretendida e a seguir calcular e mostrar o resultado.

```python
def calculadora():
    num1 = float(input("Digite o primeiro valor: "))
    num2 = float(input("Digite o segundo valor: "))

    operacao = input("Escolha a operação (+, -, *, /): ")

    if operacao == '+':
        resultado = num1 + num2
        print(f"Resultado: {num1} + {num2} = {resultado}")
    elif operacao == '-':
        resultado = num1 - num2
        print(f"Resultado: {num1} - {num2} = {resultado}")
    elif operacao == '*':
        resultado = num1 * num2
        print(f"Resultado: {num1} * {num2} = {resultado}")
    elif operacao == '/':
        if num2 != 0:
            resultado = num1 / num2
            print(f"Resultado: {num1} / {num2} = {resultado}")
        else:
            print("Erro: Divisão por zero não é permitida.")
    else:
        print("Operação inválida. Por favor, escolha entre +, -, * ou /.")

calculadora()
```

### 5. Maior e Menor de 4 Números
Leia 4 números inteiros, e exiba o maior e o menor deles.

```python
numeros = []
for i in range(4):
    while True:
        try:
            num = int(input(f"Digite o {i+1}º número inteiro: "))
            numeros.append(num)
            break
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

if numeros:
    maior_numero = max(numeros)
    menor_numero = min(numeros)
    print(f"\nO maior número é: {maior_numero}")
    print(f"O menor número é: {menor_numero}")
else:
    print("Nenhum número foi digitado.")
```
