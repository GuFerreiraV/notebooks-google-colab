### 4 - Escreva um algoritmo que leia um grupo de valores reais e determine quantos valores são positivos e quantos são negativos. Determine, também, qual é o menor desses valores. Utilize o comando de repetição que desejar.

```python
valores = []
positivos = 0
negativos = 0
menor_valor = None

print("Digite os valores reais (digite 'fim' para encerrar):")

while True:
    entrada = input()
    if entrada.lower() == 'fim':
        break
    try:
        valor = float(entrada)
        valores.append(valor)
        if valor > 0:
            positivos += 1
        elif valor < 0:
            negativos += 1

        if menor_valor is None or valor < menor_valor:
            menor_valor = valor
    except ValueError:
        print("Entrada inválida. Por favor, digite um número real ou 'fim'.")

if valores:
    print(f"\nQuantidade de valores positivos: {positivos}")
    print(f"Quantidade de valores negativos: {negativos}")
    print(f"O menor valor digitado é: {menor_valor}")
else:
    print("Nenhum valor foi digitado.")
```

### 5 - Temos um grupo de pessoas. Escreva um programa em Python que leia o sexo e a altura de cada pessoa, calcule e mostre a altura média das mulheres e dos homens separadamente. Utilize o comando de repetição que desejar

```python
alturas_homens = []
alturas_mulheres = []

print("Digite o sexo (M/F) e a altura (ex: M 1.75). Digite 'fim' para encerrar.")

while True:
    entrada = input()
    if entrada.lower() == 'fim':
        break

    partes = entrada.split()
    if len(partes) != 2:
        print("Formato inválido. Use 'M altura' ou 'F altura'.")
        continue

    sexo = partes[0].upper()
    try:
        altura = float(partes[1])
    except ValueError:
        print("Altura inválida. Por favor, digite um número.")
        continue

    if sexo == 'M':
        alturas_homens.append(altura)
    elif sexo == 'F':
        alturas_mulheres.append(altura)
    else:
        print("Sexo inválido. Use 'M' para masculino ou 'F' para feminino.")

media_altura_homens = sum(alturas_homens) / len(alturas_homens) if alturas_homens else 0
media_altura_mulheres = sum(alturas_mulheres) / len(alturas_mulheres) if alturas_mulheres else 0

print(f"\nAltura média dos homens: {media_altura_homens:.2f} m")
print(f"Altura média das mulheres: {media_altura_mulheres:.2f} m")
```

### 6 - Ler uma quantidade indeterminada de alunos com as seguintes informações RGM, NOME, Sexo e Media. Calcular a media da sala, exibir a media da sala, maior nota, menor nota e a media por sexo.

```python
alunos = []

print("Digite os dados dos alunos (RGM NOME Sexo Media). Digite 'fim' para encerrar.")
print("Exemplo: 12345 João M 8.5")

while True:
    entrada = input()
    if entrada.lower() == 'fim':
        break

    partes = entrada.split()
    if len(partes) < 4:
        print("Formato inválido. São necessários RGM, NOME, Sexo e Media.")
        continue

    try:
        rgm = partes[0]
        nome = partes[1]
        sexo = partes[2].upper()
        media = float(partes[3])

        if sexo not in ['M', 'F']:
            print("Sexo inválido. Use 'M' para masculino ou 'F' para feminino.")
            continue

        alunos.append({'RGM': rgm, 'NOME': nome, 'SEXO': sexo, 'MEDIA': media})
    except ValueError:
        print("Média inválida. Por favor, digite um número.")
    except Exception as e:
        print(f"Erro ao processar a entrada: {e}")

if alunos:
    total_medias = sum(aluno['MEDIA'] for aluno in alunos)
    media_sala = total_medias / len(alunos)

    maior_nota = max(aluno['MEDIA'] for aluno in alunos)
    menor_nota = min(aluno['MEDIA'] for aluno in alunos)

    medias_masculino = [aluno['MEDIA'] for aluno in alunos if aluno['SEXO'] == 'M']
    medias_feminino = [aluno['MEDIA'] for aluno in alunos if aluno['SEXO'] == 'F']

    media_masculino = sum(medias_masculino) / len(medias_masculino) if medias_masculino else 0
    media_feminino = sum(medias_feminino) / len(medias_feminino) if medias_feminino else 0

    print(f"\n--- Relatório de Alunos ---")
    print(f"Média da sala: {media_sala:.2f}")
    print(f"Maior nota: {maior_nota:.2f}")
    print(f"Menor nota: {menor_nota:.2f}")
    print(f"Média das notas masculinas: {media_masculino:.2f}")
    print(f"Média das notas femininas: {media_feminino:.2f}")
else:
    print("Nenhum aluno foi inserido.")
```

### 7 - Ler vários produtos (código, descrição, quantidade e valor) para uma venda, exibir a lista de produtos e o total da venda.

```python
produtos_venda = []
total_venda = 0

print("Digite os dados dos produtos (Código Descrição Quantidade Valor). Digite 'fim' para encerrar.")
print("Exemplo: P001 Caneta 5 2.50")

while True:
    entrada = input()
    if entrada.lower() == 'fim':
        break

    partes = entrada.split()
    if len(partes) < 4:
        print("Formato inválido. São necessários Código, Descrição, Quantidade e Valor.")
        continue

    try:
        codigo = partes[0]
        descricao = partes[1]
        quantidade = int(partes[2])
        valor_unitario = float(partes[3])

        if quantidade <= 0 or valor_unitario < 0:
            print("Quantidade deve ser positiva e valor não pode ser negativo.")
            continue

        valor_total_item = quantidade * valor_unitario
        produtos_venda.append({
            'Código': codigo,
            'Descrição': descricao,
            'Quantidade': quantidade,
            'Valor Unitário': valor_unitario,
            'Total Item': valor_total_item
        })
        total_venda += valor_total_item
    except ValueError:
        print("Quantidade ou Valor inválido. Por favor, digite números.")
    except Exception as e:
        print(f"Erro ao processar a entrada: {e}")

if produtos_venda:
    print("\n--- Lista de Produtos da Venda ---")
    for produto in produtos_venda:
        print(f"Código: {produto['Código']}, Descrição: {produto['Descrição']}, "
              f"Qtd: {produto['Quantidade']}, Valor Unit.: {produto['Valor Unitário']:.2f}, "
              f"Total Item: {produto['Total Item']:.2f}")
    print(f"\nTotal da Venda: R$ {total_venda:.2f}")
else:
    print("Nenhum produto foi adicionado à venda.")
```
