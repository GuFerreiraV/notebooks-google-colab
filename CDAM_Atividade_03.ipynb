**Importante!!!**

De maneira geral, dados faltantes devem ser considerados caso a caso. Muitas vezes podem na verade ter um significado específico no contexto da coleta dos dados.

Quando a informação de uma (ou mais) variáveis não estiver disponível (por exemplo: falta de resposta, rasura na marcação da resposta, ou simplesmente não foi possível coletar aquela informação para aquele indivíduo), esta informação deve ser marcada (ou indicada) com algum símbolo especial ou permanecer em branco, o que pode resultar em valores 'null' ou 'NaN' na importação dos dados em versão eletrônica.

Não é SEMPRE que esses casos devem ser preenchidos com 0 (zero), pois o zero não significa falta de informação, e sim uma quantidade que foi medida, mas foi verificada como nula.

No caso da coluna filhos, atribuir ou não valores 0 para valores desconhecidos 'NaN' pode, por exemplo, impactar o cálculo de uma média de filhos por funcionário.

Para fins didáticos faremos a substituição dos valores 'NaN' por 'zero', considerando por exemplo que o formulário que gerou estes dados cotinha uma opção de 'Não se aplica' ou 'Não tenho filhos', gerando valores não numérios que ao serem 'convertidos' geraram entradas 'NaN'.

<font color='red'>**Exercício:**</font> Substitua os valores 'NaN' por 'zero' da coluna **Filhos** por zero. Utilize `np.nan` para encontrar valores desconhecidos, e em seguida mostre o resultado.

<font color='red'>**Exercício:**</font> Gere um gráfico de setores (pizza) do atributo **Estado Civil** de forma que seja possível ver as porcentagens com 1 casa decimal, e a classe **solteiro** apareça em detaque (separada), incluindo sombra sombra.

Consulte a documentação da função [pd.plot.pie()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.pie.html) para gerar o gráfico.

Consulte a documentação dos atributos do **gráfico de setores**:
[Pie chart features](https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html)

```python
estado_civil_counts = df['Est.civil'].value_counts()

try:
    solteiro_index = list(estado_civil_counts.index).index('Solteiro')
    explode_list = [0] * len(estado_civil_counts)
    explode_list[solteiro_index] = 0.1
except ValueError:
    explode_list = [0] * len(estado_civil_counts)

plt.figure(figsize=(8, 8))
estado_civil_counts.plot.pie(
    autopct='%.1f%%',
    startangle=90,
    explode=explode_list,
    shadow=True
)

plt.title('Estado Civil')
plt.ylabel('')
plt.show()
```

### Análise Bivariada

Na análise bivariada procuramos identificar relações entre duas variáveis. Assim como na análise univariada, estas relações podem ser resumidas por gráficos, tabelas e/ou medidas estatísticas. O tipo de resumo vai depender dos tipos das variáveis envolvidas

<font color='red'>**Exercício:**</font> Modifique o exemplo abaixo para gerar gráficos comparativos entre as variáveis qualitativas **Est.Civil** e **Instrucao** como os dos exemplos a seguir:

Consulte a documentação da função [pd.plot.bar()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html) para gerar o gráfico.

Extra: aproveite para tentar utilizar o recurso de subplot do matplotlib para colocar 4 gráficos em uma única imagem



```python
# 2. Crie a figura com 4 subplots (2 linhas, 2 colunas)
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))
fig.suptitle('Análise Comparativa de Estado Civil e Instrução', fontsize=18, y=0.95)

# 3. Gráfico 1 (Superior Esquerdo): Barras empilhadas por Instrucao
ax1 = axs[0, 0]
tabela_instrucao_ec = pd.crosstab(df['Inst'], df['Est.civil'])
tabela_instrucao_ec.plot.bar(stacked=True, ax=ax1, rot=0)
ax1.set_title('Instrução por Estado Civil')
ax1.set_xlabel('Instrução')
ax1.set_ylabel('Contagem')

# 4. Gráfico 2 (Superior Direito): Barras empilhadas por Estado Civil
ax2 = axs[0, 1]
tabela_ec_instrucao = pd.crosstab(df['Est.civil'], df['Inst'])
tabela_ec_instrucao.plot.bar(stacked=True, ax=ax2, rot=0)
ax2.set_title('Estado Civil por Instrução')
ax2.set_xlabel('Estado Civil')
ax2.set_ylabel('Contagem')

# 5. Gráfico 3 (Inferior Esquerdo): Barras lado a lado por Instrucao
ax3 = axs[1, 0]
tabela_instrucao_ec.plot.bar(ax=ax3, rot=0)
ax3.set_title('Instrução por Estado Civil')
ax3.set_xlabel('Instrução')
ax3.set_ylabel('Contagem')

# 6. Gráfico 4 (Inferior Direito): Barras lado a lado normalizadas
ax4 = axs[1, 1]
tabela_normalizada = pd.crosstab(df['Est.civil'], df['Inst'], normalize='index')
tabela_normalizada.plot.bar(ax=ax4, rot=0)
ax4.set_title('Proporção de Instrução por Estado Civil')
ax4.set_xlabel('Estado Civil')
ax4.set_ylabel('Proporção')

# 7. Ajuste o layout para evitar sobreposição de títulos e rótulos
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
